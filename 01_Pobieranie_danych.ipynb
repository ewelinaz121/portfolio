Retrieving data from the `airport` endpoint


```python
# Initialization of an empty list with airport data
airport_data = []

# Iterating through unique airport IDs from a CSV file
for airport_id in origin_airport_ids:
    print(f"Wysyłanie zapytania dla ID lotniska: {airport_id}")
    # Tworzenie adresu URL dla danego lotniska
    url = f"https://api-datalab.coderslab.com/api/v2/airport/{airport_id}"
    print("URL zapytania:", url)
    
    # Sending an HTTP GET request to the API
    response = requests.get(url, headers=headers)
    
    # Check if the answer is correct (code 200)
    if response.status_code == 200:
        print("Odpowiedź otrzymana pomyślnie.")
        # Convert data to JSON format
        data = response.json()
        
        # Adding airport details to the listy
        airport_data.append(data)
    else:
        # Printing a message about missing data for a given airport ID
        print(f"Dla ID {airport_id} brak danych. Kod błędu: {response.status_code}")

# Converting an airport data list to a Pandas DataFrame object
airport_df = pd.DataFrame(airport_data)
```

Retrieving data from the airport weather endpoint

```python
def call_airport_weather_API(date):
    response = requests.get(f'https://api-datalab.coderslab.com/api/v2/airportWeather?date={date}', headers=headers)
    if response.ok == True:
        return response.json() 
    else:
        print(f"Bad request for date {date}") 
        return None
```

```python
# Initialization of an empty list with airport weather data
airport_weather_data = []

# Specifying the start date
start_date = dt.datetime(2019, 1, 1)

# Iterate for 15 months
for i in range(15):
    # Downloading weather data for a given month
    weather_report = call_airport_weather_API(start_date.strftime('%Y-%m'))
    if weather_report:
        airport_weather_data.extend(weather_report)
    print(dt.datetime.now(), " Pobrane dane dla miesiąca: ", start_date.strftime('%Y-%m'), f" Obieg pętli: {i}", " Pobrane dane dla miesiąca: ", start_date.strftime('%Y-%m'), f" Obieg pętli: {i}")  # Wyświetlenie logów
    start_date += relativedelta(months=1)

# Convert a list of weather data to a Pandas data frame
airport_weather_df = pd.DataFrame(airport_weather_data)
```
